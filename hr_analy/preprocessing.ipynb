## Dealing with Distributions

### Correlation Chart

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# 1. Convert Attrition to numeric (if not already numeric)
if not pd.api.types.is_numeric_dtype(df['Attrition']):
    df['Attrition'] = df['Attrition'].map({'Yes': 1, 'No': 0})

# 2. Select relevant columns + Attrition
cols_to_check = [
    'Age', 'DailyRate', 'DistanceFromHome', 'EmployeeNumber', 'HourlyRate',
    'MonthlyRate', 'PercentSalaryHike', 'TotalWorkingYears', 
    'YearsAtCompany', 'YearsInCurrentRole', 'YearsSinceLastPromotion', 
    'YearsWithCurrManager'
]

# Keep only the columns that actually exist in df
cols_present = [col for col in cols_to_check if col in df.columns]
selected_df = df[['Attrition'] + cols_present].dropna()

# 3. Calculate correlations with Attrition
correlations = selected_df.corr()['Attrition'].drop('Attrition').sort_values(ascending=False)

# 4. Print correlation values
print("Correlation of Selected Features with Attrition:\n")
print(correlations)

# 5. Visualize as a bar chart
plt.figure(figsize=(8,6))
sns.barplot(x=correlations.values, y=correlations.index, palette="coolwarm")
plt.title("Correlation of Numerical Features with Attrition", fontsize=14)
plt.xlabel("Correlation Coefficient")
plt.ylabel("Numerical Features")
plt.show()
```

#### This plot is showing the correlation of Attrition with Other Numerical Columns

No Transformation Needed. Binning this column into categories (0–3 years, 4–10 years, 11–20 years, 20+ years) to find which tenure group has highest attrition.

```python
df['YearsAtCompany'].unique()
```

```python
import pandas as pd

# Define bins and labels
bins = [0, 3, 10, 20, float('inf')]
labels = ['0-3 years', '4-10 years', '11-20 years', '20+ years']

# Replace the column with binned categories
df['YearsAtCompany'] = pd.cut(df['YearsAtCompany'], 
                              bins=bins, 
                              labels=labels, 
                              right=True, 
                              include_lowest=True)

# Verify
print(df['YearsAtCompany'].value_counts())
```

YearsAtCompany has been converted into Category
